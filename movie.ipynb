```python
movie = movie[["title", "id" , "tagline" , "genres"   , "keywords"  , "overview"  , "cast" , "crew"]]
```

```python
movie.head()
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])

    return L
```

```python
movie["genres"] = movie["genres"].apply(convert)
```

```python
movie.head()
```

```python
movie["keywords"] = movie["keywords"].apply(convert)
```

```python
movie.head(1)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
    return L
```

```python
def convert3(obj):
    L = []
    cnt=0
    for i in ast.literal_eval(obj):
        if cnt != 3:
            L.append(i["name"])
            cnt+=1
        else:
            break
        
    return L
```

```python
movie["cast"] = movie["cast"].apply(convert3)
```

```python
movie["crew"] = movie["crew"].apply(fetch_director)
```

```python
movie.head(2)
```

```python
movie["cast"] = movie["cast"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movie["genres"] = movie["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
movie["keywords"] = movie["keywords"].apply(lambda x: [i.replace(" ", "") for i in x])
movie["crew"] = movie["crew"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movie.head(2)
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie["tags"] = movie["overview"] + movie["genres"] + movie["keywords"] + movie["cast"] + movie["crew"]
```

```python
movie.head(1)
```

```python
movie["tags"]
```

```python
new = movie.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new.head(2)
```

```python
new["tags"] = new["tags"].apply(lambda x: " ".join(x)) # converting the list into string
```

```python
new.head()
```

```python
new["tags"] = new["tags"].apply(lambda x: x.lower())
```

```python
new.head()
```

### `Removing similar texts using stemming `

```python
import nltk
```

```python
# This is used to find the root form of the word. Extracts single word from the repeating words.
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    t = []
    
    for i in text.split():
        t.append(ps.stem(i))
    
    return " ".join(t)   # This is used to join the list elements into a single string.
```

```python
new["tags"] = new["tags"].apply(stem)
```

```python
new["tags"][0]
```

```python
new["tags"][1]
```

## `Vectorization`

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new["tags"]).toarray()
```

```python
vectors
```

```python
vectors[0].shape
```

```python
for i in cv.get_feature_names_out():
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity[0] # similarity of 1st movie with 1st will be 1
```

```python
similarity[1] # similarity of 2st movie with 2nd will be 1
```

```python
def recommend(movie):
    movie_index = new[new["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new.iloc[i[0]].title)
```
